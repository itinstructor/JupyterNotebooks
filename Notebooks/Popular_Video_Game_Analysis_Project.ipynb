Creating a column and dropping them.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Create a dataframe with new edited format without TBA strings
df = pd.read_csv("https://raw.githubusercontent.com/itinstructor/JupyterNotebooks/main/Datasets/Wilson/games1.csv")
# Create column with only Release year
df['Release Year'] = pd.to_datetime(df['Release Date']).dt.strftime('%Y')
# Change data type
df['Release Year'] = df['Release Year'].astype('int')
# Drop unnecessary columns
df = df.drop(['Release Date','Playing1'], axis=1)
display(df)
```

```python
# Sort Release year
df = df.sort_values(by='Release Year', ascending=False)
# Sort by specific years
df = df.loc[df['Release Year'].between(2020,2023)]
# Sort Playing column
df = df.sort_values(by='Playing', ascending=False)
# Drop duplicated titles 
df = df.drop_duplicates(subset='Title')
# Only have top 10 most "playing" games
df = df.head(10)

```

```python
# Set size of plot 
plt.rcParams["figure.figsize"] = (10, 5)
# Create bar plot
df.plot(kind='bar', x='Title', y='Playing')
plt.grid(axis = 'y')
plt.show()
```
