What is the most popular halloween candy ingredent flavor?

Let's use the candy-data database from https://github.com/fivethirtyeight/data/blob/master/candy-power-ranking/candy-data.csv

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/candy-power-ranking/candy-data.csv')

# display info and staticstics
print(df.info())
display(df.describe())

display(df)
```

I will analyze the data by adding up the ingredent flavors use and displaying it in a bar plot

```python
import matplotlib.pyplot as plt
import pandas as pd
chocolate = 0
fruity = 0
caramel = 0
peanutyalmondy = 0
nougat = 0
crispedricewafer = 0

# Add up each use of ingredents
for index, row in df.iterrows():
  if row['chocolate'] == 1:
    chocolate = chocolate + 1
  if row['fruity'] == 1:
    fruity = fruity + 1
  if row['caramel'] == 1:
    caramel = caramel + 1
  if row['peanutyalmondy'] == 1:
    peanutyalmondy = peanutyalmondy + 1
  if row['nougat'] == 1:
    nougat = nougat + 1
  if row['crispedricewafer'] == 1:
    crispedricewafer = crispedricewafer + 1

dataframe = pd.DataFrame({'name':['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer'],
                          'ingredent_use':[chocolate, fruity, caramel, peanutyalmondy, nougat, crispedricewafer]})

dataframe.plot(kind='bar', x='name', y='ingredent_use')
plt.show()
```

In conclusion the most used ingredent flavor is fruity with chocolate just behind.
