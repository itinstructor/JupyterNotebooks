```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import urllib.request
from skimage import io

# URL of the image
# url = "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg"
# url = "https://wncc.edu/_resources/assets/images/footer-1400@1x.jpg"
# url = "https://d2o2figo6ddd0g.cloudfront.net/6/c/267jqi57ap6brk/Luna_--_game_1_Saturday_8x10-7048.jpg"
url = "https://upload.wikimedia.org/wikipedia/en/thumb/1/1e/Baseball_%28crop%29.jpg/1200px-Baseball_%28crop%29.jpg"

# Fetch the image data from the URL
img = io.imread(url)

# Decode the image data and load it with OpenCV
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using OpenCV
cv2_imshow(img)
```

#Convert Image to Grayscale

```python
# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the grayscale image using OpenCV
cv2_imshow(gray_img)
```

# Guassian Blur

```python
# Apply Gaussian blur to the grayscale image
blurred_img = cv2.GaussianBlur(gray_img, (5, 5), 0)

# Display the blurred image
cv2_imshow(blurred_img)
```

#Edge Detection using Canny

```python
# Apply Canny edge detection to the blurred image
edges = cv2.Canny(blurred_img, 50, 150)
# Display the blurred image
cv2_imshow(edges)
```
